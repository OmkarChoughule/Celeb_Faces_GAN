```python
with tf.device('/GPU:0'):
    directory = 'archive\img_align_celeba'

    # Get all images from dir
    dataset = tf.keras.preprocessing.image_dataset_from_directory(directory=directory, color_mode='rgb',
                                                   batch_size=128, image_size=(32,32), label_mode=None,
                                                   shuffle=True, seed=42)

    # Normalize the data

    dataset = dataset.map(lambda x: (x / 127.5) - 1)  # Normalize to [-1,1]
```

GANs-> Generartive Adversarial Network
Generator, Disc.
Disc.-> Image Classification Task
Latent-> Noise

```python
from tensorflow.keras.layers import Conv2D, LeakyReLU,BatchNormalization, Dropout, Flatten, Dense, Activation, Reshape
from tensorflow.keras import Model, Input

# Discriminator Model
def discriminator_model():
    disc_input = Input(shape=(32, 32, 3), name='discriminator_network')
    x = Conv2D(64, kernel_size=3, strides=2, padding='same')(disc_input)
    x = LeakyReLU(0.2)(x)
    x = Dropout(0.3)(x)

    x = Conv2D(128, kernel_size=3, strides=2, padding='same')(x)
    x = LeakyReLU(0.2)(x)
    x = Dropout(0.3)(x)

    x = Flatten()(x)
    x = Dense(1, activation='sigmoid')(x)  # Keep sigmoid for BCE loss

    disc = Model(disc_input, x)
    return disc
```

```python
with tf.device('/GPU:0'):
    disc_model = discriminator_model()
disc_model.summary()
```

```python
from tensorflow.keras.layers import Conv2D, LeakyReLU,BatchNormalization, Dropout, Flatten, Dense, Activation, Reshape, Conv2DTranspose
from tensorflow.keras import Model, Input
# Generator Model
def generator_model(z_dim=100):
    gen_input = Input(shape=(z_dim,), name='generator_network')

    x = Dense(8 * 8 * 128)(gen_input)
    x = Reshape((8, 8, 128))(x)
    x = BatchNormalization()(x)
    x = LeakyReLU(0.2)(x)

    x = Conv2DTranspose(128, kernel_size=4, strides=2, padding='same')(x)
    x = BatchNormalization()(x)
    x = LeakyReLU(0.2)(x)

    x = Conv2DTranspose(64, kernel_size=4, strides=2, padding='same')(x)
    x = BatchNormalization()(x)
    x = LeakyReLU(0.2)(x)

    x = Conv2DTranspose(3, kernel_size=3, strides=1, padding='same', activation='tanh')(x)  # tanh output
    
    gen = Model(gen_input, x)
    return gen
```

```python
with tf.device('/GPU:0'):
    gen_model = generator_model()
```

```python
gen_model.summary(0)
```

```python
disc_model = discriminator_model()
gen_model = generator_model()
# Image Classification CNN
# Compile the model that trains disc.
disc_model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.0002, beta_1=0.5), 
                   loss='binary_crossentropy', 
                   metrics=['accuracy'])
```

```python
# Compile the model that trains generator
z_dim = 100
disc_model.trainable = False
model_input = Input(shape=(z_dim,))
model_output = disc_model(gen_model(model_input))
model = Model(model_input, model_output)
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.0002, beta_1=0.5), 
              loss='binary_crossentropy')
```

```python
model.summary()
```

```python
# Training Functions
def train_discriminator(batch, batch_size):
    real_labels = np.ones((batch_size, 1)) * 0.9  # Smooth labels
    fake_labels = np.zeros((batch_size, 1)) + 0.1

    disc_model.train_on_batch(batch, real_labels)

    noise = np.random.normal(0, 1, (batch_size, z_dim))
    gen_images = gen_model.predict(noise)

    disc_model.train_on_batch(gen_images, fake_labels)

def train_generator(batch_size):
    valid_labels = np.ones((batch_size, 1))
    noise = np.random.normal(0, 1, (batch_size, z_dim))
    model.train_on_batch(noise, valid_labels)
```

```python
def save_image(fake_image, path):
    image = ((fake_image + 1) / 2 * 255).astype(np.uint8)  # Convert back to [0,255]
    tf.keras.preprocessing.image.save_img(path, image)
    return image

def generate_and_save_image(path):
    noise = np.random.normal(0, 1, (1, z_dim))
    fake_images = gen_model(noise)
    image = save_image(fake_images[0].numpy(), path)
    plt.imshow(image)
    plt.show()
```

```python
# tf.config.experimental.set_memory_growth(tf.config.list_physical_devices('GPU')[0], True)
```

```python
# Training Loop
PATH = 'generate_face'
os.makedirs(PATH, exist_ok=True)
num_epochs = 100

for epoch in range(num_epochs):
    print(f'Epoch {epoch + 1}')
    for i, batch in enumerate(dataset):
        train_discriminator(batch, batch.shape[0])
        train_generator(batch.shape[0])

        if i % 100 == 0:
            print(f'Batch {i}')
            generate_and_save_image(f'{PATH}/epoch_{epoch}_batch_{i}.jpg')
            
```
